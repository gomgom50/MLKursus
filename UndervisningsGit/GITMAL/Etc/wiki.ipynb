### Testing for GPUs

Look for inspiration in the python file 

```libitmal/kernelfuns.py```
  
and try to call the function 

```StartupSequence_GPU(verbose=True)```

It this still fails, try to create a function, that checks for GPUs ala:

````python
import tensorflow

def MyGPUCheck():
    print("MyGPUCheck():..")
    physical_devices = tensorflow.config.list_physical_devices('GPU')
    n = len(physical_devices)
    print(f"  found {n} GPU device(s)")    
    for i in physical_devices:
        print(f"  {i}")
    print("OK")
    
MyGPUCheck()
````

A succefull call to thelist of physical devices may output something like

```

  PhysicalDevice(name='/physical_device:GPU:0', device_type='GPU')
OK
```

### Swiching between Jupyter Notebook or JupyterLab

You can switch between Notebooks and JupyterLab in the browser url/link adress 

Notebooks urls look like:

```
http://localhost:8888/tree
http://localhost:8888/notebooks/GITMAL/Etc/wiki2.ipynb
```
where you can change `tree` or `notebooks` to `lab` for running JupyterLab instead

```
http://localhost:8888/lab
```

The Lab enviroment is more advanced that the Notebook, for example it has a buildin debug view, whereas Notebook only have a debugger if you install `nbextensions`, see  <a url=https://jupyter-contrib-nbextensions.readthedocs.io/en/latest/install.html>jupyter-contrib-nbextensions </a>

## SSL Error

Sometimes conda fail downloading with an SSL error. At one point in time the IT-security course were known to fiddle with the Windows SSL certificates, and was sometimes causing the error.

In this case, no one has changed anything on the PC regarding SSL certificates, and the course of the problem is unknown.



FIX: just update conda in the base enviroment ala  

```
> conda activate base
> conad update --all
```

The error happens for an non-updated base conda bin and when trying to use it in a newly created environemt, ala the freshly created `(swmal)` environment in the screenshot. Only seen in the Windows OS.
